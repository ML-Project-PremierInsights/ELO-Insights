```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
match_data = pd.read_csv('../Datasets/premier-league-matches.csv')
team = 'Manchester Utd'
```

```python
home_team_wins = len(match_data[(match_data['Home'] == team) & (match_data['FTR'] == 'H')])
home_team_losses = len(match_data[(match_data['Home'] == team) & (match_data['FTR'] == 'A')])
home_team_draws = len(match_data[(match_data['Home'] == team) & (match_data['FTR'] == 'D')])

away_team_wins = len(match_data[(match_data['Away'] == team) & (match_data['FTR'] == 'A')])
away_team_losses = len(match_data[(match_data['Away'] == team) & (match_data['FTR'] == 'H')])
away_team_draws = len(match_data[(match_data['Away'] == team) & (match_data['FTR'] == 'D')])
```

```python
colors = ['#4CAF50', '#F44336', '#FFEB3B']

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Match Result Proportions for Manchester Utd', fontsize=16, fontweight='bold')

# Home team pie chart
axs[0].pie([home_team_wins, home_team_losses, home_team_draws], 
           labels=['Wins', 'Losses', 'Draws'], 
           autopct='%1.1f%%', startangle=90, colors=colors, textprops={'fontsize': 12, 'weight': 'bold'})
axs[0].set_title('Home Games Proportion', fontsize=14, fontweight='bold')

# Away team pie chart
axs[1].pie([away_team_wins, away_team_losses, away_team_draws], 
           labels=['Wins', 'Losses', 'Draws'], 
           autopct='%1.1f%%', startangle=90, colors=colors, textprops={'fontsize': 12, 'weight': 'bold'})
axs[1].set_title('Away Games Proportion', fontsize=14, fontweight='bold')

# Total games pie chart
axs[2].pie([home_team_wins + away_team_wins, home_team_losses + away_team_losses, home_team_draws + away_team_draws], 
           labels=['Wins', 'Losses', 'Draws'], 
           autopct='%1.1f%%', startangle=90, colors=colors, textprops={'fontsize': 12, 'weight': 'bold'})
axs[2].set_title('Total Games Proportion', fontsize=14, fontweight='bold')

# Tighten layout and display the plots
plt.tight_layout()
plt.subplots_adjust(top=0.85)
plt.show()
```
