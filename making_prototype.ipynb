```python
fixtures = pd.read_csv('Datasets/fixtures-2025.csv')
fixtures.head()

teams = set(list(fixtures['home']) + list(fixtures['away']))
teams = list(teams)
print(teams)
print(len(teams))
```

```python
team_data = {}
teams = ['Chelsea', 'Everton', 'Liverpool', 'Ipswich Town', 'Leicester City', 'Manchester Utd', 'Newcastle Utd', 'Bournemouth', 'Manchester City', 'Tottenham', "Nott'ham Forest", 'Brighton', 'Brentford', 'Fulham', 'West Ham', 'Wolves', 'Aston Villa', 'Southampton', 'Crystal Palace', 'Arsenal']
```

```python
for i, row in dataset.iterrows():
    home_team = row['Home Team']
    away_team = row['Away Team']

    if home_team in teams:
        if home_team not in team_data:
            team_data[home_team] = {
                'ELO':row['Home Team ELO'], 
                'XG':row['Home XG'],
                'XGA':row['Home XGA'],
                'Win Percentage':row['Home Win Percentage'], 
                'Home Draw Percentage':row['Home Draw Percentage'],
                'Team Form':row['Home Team Form'],
                'Team Cumulative Points':row['Home Team Cumulative Points']
            }
        else:
            team_data[home_team]['ELO'] = row['Home Team ELO']
            team_data[home_team]['XG'] = row['Home XG']
            team_data[home_team]['XGA'] = row['Home XGA']
            team_data[home_team]['Win Percentage'] = row['Home Win Percentage']
            team_data[home_team]['Home Draw Percentage'] = row['Home Draw Percentage']
            team_data[home_team]['Team Form'] = row['Home Team Form']
            team_data[home_team]['Team Cumulative Points'] = row['Home Team Cumulative Points']

    if away_team in teams:
        if away_team not in team_data:
            team_data[away_team] = {
                'ELO':row['Away Team ELO'], 
                'XG':row['Away XG'],
                'XGA':row['Away XGA'],
                'Win Percentage':row['Away Win Percentage'], 
                'Away Draw Percentage':row['Away Draw Percentage'],
                'Team Form':row['Away Team Form'],
                'Team Cumulative Points':row['Away Team Cumulative Points']
            }
        else:
            team_data[away_team]['ELO'] = row['Away Team ELO']
            team_data[away_team]['XG'] = row['Away XG']
            team_data[away_team]['XGA'] = row['Away XGA']
            team_data[away_team]['Win Percentage'] = row['Away Win Percentage']
            team_data[away_team]['Away Draw Percentage'] = row['Away Draw Percentage']
            team_data[away_team]['Team Form'] = row['Away Team Form']
            team_data[away_team]['Team Cumulative Points'] = row['Away Team Cumulative Points']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score,confusion_matrix, mean_squared_error, r2_score, precision_score, recall_score, log_loss
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

print(dataset.columns)
print(dataset.shape)
X_train, X_test, y_train, y_test = train_test_split(dataset[['Home Team ELO', 'Home XG', 'Home XGA', 'Home Win Percentage', 'Home Draw Percentage', 'Home Team Form', 'Home Team Cumulative Points', 'Away Team ELO', 'Away XG', 'Away XGA', 'Away Win Percentage', 'Away Draw Percentage', 'Away Team Form', 'Away Team Cumulative Points']], dataset['Winner'], test_size=0.2, random_state=42)
random_forest_model = RandomForestClassifier(random_state=16, n_estimators=1500)

random_forest_model.fit(X_train, y_train)

y_pred_rf = random_forest_model.predict(X_test)

accuracy_rf = accuracy_score(y_test, y_pred_rf)
print('Random Forest Accuracy:', accuracy_rf)
```

```python
team_data = pd.read_csv('Datasets/2023-2024_team_data.csv')
print(team_data.head())
```

```python
final_standings = {team:0 for team in team_data['Team']}
print(final_standings)
```

```python
for i, row in fixtures.iterrows():
    home_team = row['home']
    away_team = row['away']
    home_ELO = team_data[team_data['Team'] == home_team]['ELO'].values[0]
    home_XG = team_data[team_data['Team'] == home_team]['XG'].values[0]
    home_XGA = team_data[team_data['Team'] == home_team]['XGA'].values[0]
    home_win_percentage = team_data[team_data['Team'] == home_team]['Win Percentage'].values[0]
    home_draw_percentage = team_data[team_data['Team'] == home_team]['Draw Percentage'].values[0]
    home_team_form = team_data[team_data['Team'] == home_team]['Team Form'].values[0]
    home_team_cumulative_points = team_data[team_data['Team'] == home_team]['Team Cumulative Points'].values[0]

    away_ELO = team_data[team_data['Team'] == away_team]['ELO'].values[0]
    away_XG = team_data[team_data['Team'] == away_team]['XG'].values[0]
    away_XGA = team_data[team_data['Team'] == away_team]['XGA'].values[0]
    away_win_percentage = team_data[team_data['Team'] == away_team]['Win Percentage'].values[0]
    away_draw_percentage = team_data[team_data['Team'] == away_team]['Draw Percentage'].values[0]
    away_team_form = team_data[team_data['Team'] == away_team]['Team Form'].values[0]
    away_team_cumulative_points = team_data[team_data['Team'] == away_team]['Team Cumulative Points'].values[0]

    entry = [home_ELO, home_XG, home_XGA, home_win_percentage, home_draw_percentage, home_team_form, home_team_cumulative_points, away_ELO, away_XG, away_XGA, away_win_percentage, away_draw_percentage, away_team_form, away_team_cumulative_points]
    entry_df = pd.DataFrame([entry], columns=X_train.columns)

    predicted_result = random_forest_model.predict(entry_df)
    print(f'{home_team} vs {away_team}: {predicted_result[0]}')

    if predicted_result[0] == 1:
        final_standings[home_team] += 3

    elif predicted_result[0] == 0:
        final_standings[home_team] += 1
        final_standings[away_team] += 1

    elif predicted_result[0] == -1:
        final_standings[away_team] += 3

    


    
```

```python
final_standings = dict(sorted(final_standings.items(), key=lambda item: item[1], reverse=True))
for team in final_standings:
    print(f'{team}: {final_standings[team]}')
```
