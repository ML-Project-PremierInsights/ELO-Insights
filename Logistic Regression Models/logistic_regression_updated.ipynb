```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('/Users/noeltiju/Desktop/ML-Project/ELO-Insights/updated_training_data.csv')

# Reset the index to convert the MultiIndex into columns
dataset = dataset.reset_index()
col_names = [
    'Home Team ELO',
    'Away Team ELO',
    'Home Team Win Percentage',
    'Home Team Draw Percentage',
    'Away Team Win Percentage',
    'Away Team Draw Percentage',
    'Winner'
]

dataset = dataset[col_names]
print(dataset.head())
```

```python
features = ['Home Team ELO','Away Team ELO', 'Home Team Win Percentage', 'Home Team Draw Percentage', 'Away Team Win Percentage', 'Away Team Draw Percentage']
target = 'Winner'

X = dataset[features]
y = dataset[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=16)
```

```python
logistic_regression_model = LogisticRegression(random_state=16, max_iter=1000, solver='saga')


# fit the model with data
logistic_regression_model.fit(X_train, y_train)

y_pred = logistic_regression_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)

import matplotlib.pyplot as plt
import seaborn as sns

# Plot the confusion matrix
class_labels = ['Away Team Win', 'Draw', 'Home Team Win']

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cnf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
